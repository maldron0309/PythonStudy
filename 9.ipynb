도도새는 1분에 하나씩 먹어요. 그 먹은 밥 이름이 food 안에 저장이 되어있어요. 처음 밥을 먹었을 때는 0분이라고 할 때 도도새가 복통을 일으킨 시점은 얼마나 시간이 흐른 후였을까요?

- 도도새가 복통이 난 시점을 계산하여 time 변수에 저장해 주세요.

**Tip!**
- 예시는 예시일 뿐입니다. 모든 리스트에 대해서 올바른 시점을 계산할 수 있도록 코드를 작성해봅시다.

```
# 도도새가 언제 배탈이 났는지 계산해서 리턴해 보세요.
def vomit(food):
    # 도도새가 배탈 난 시점을 time 변수에 저장해 주세요.
    time = 0

    # food 리스트를 반복하면서 웩이라고 비명을 질렀는지 확인합니다.
    for i in food:
        if i == '웩':
            break
        time += 1

    return time

food = ['냠', '냠', '쩝', '쩝', '땡', '웩', '아', '이', '고', '내', '배', '야']

print(vomit(food))
```

### 수가 든 상자


**수가 든 상자**

엘리스의 눈앞에는 수를 넣을 수 있는 상자 number_box 가 있습니다.

number_box는 언제든지 자유롭게 수를 넣을 수 있게 되어있습니다.

하지만 수를 빼낼 때는 상자 안에 들어 있는 수 중에서 가장 큰 수만 빼낼 수 있게 되어있습니다.

number_box를 직접 구현해보도록 합시다.

>     #클래스

**스스로 해봐요!**

다음 기능을 가지고 있는 number_box()클래스를 구현해 봅시다.

- 필드 num: 숫자를 담고 있는 리스트입니다.

- 메서드 push(n) : n을 추가합니다. 해당 수는 num안에 저장됩니다.

- 메서드 pop() : 저장되어 있는 수 중에서 가장 큰 수를 반환합니다. 그 후에 반환한 수를 제거합니다.

```
class number_box():
    # number_box()에 필요한 필드, 메서드를 구현해주세요.
    def __init__(self):
        self.num = []

    def push(self, n):
        self.num.append(n)

    def pop(self):
        if not self.num:
            return None
        max_num = max(self.num)
        self.num.remove(max_num)
        return max_num

box = number_box()
box.push(5)
box.push(3)
box.push(9)

print(box.pop())
print(box.pop())
print(box.pop())
print(box.pop())
```

### 냉정과 열정 사이


**냉정과 열정 사이에 있는 사랑의 개수는?**  
문자열 “냉정”, “열정”이 각각 1개씩, 나머지는 모두 문자열 “사랑”으로 가득 찬 리스트 emotion이 주어질 때, 냉정과 열정 사이의 “사랑” 문자열이 이 얼마나 있는지 반환하는 함수 hot_cold()를 작성해주세요.

    #반복문 #조건문

**스스로 해봐요!**.
- hot_cold함수에는 인자로 리스트 emotion이 주어집니다.
- emotion에는 ‘냉정’과 ‘열정’이 1개씩 있습니다.
- 이 둘 사이의 문자열 ‘사랑’ 의 개수를 세어 반환해주세요!

```
def hot_cold(emotion):
    # emotion 리스트 속에서 냉정과 열정 사이의 사랑의 개수를 세어 반환해 보세요.
    count_love = 0
    start_index = emotion.index('냉정')
    end_index = emotion.index('열정')

    for i in range(start_index + 1, end_index):
        if emotion[i] == '사랑':
            count_love += 1

    return count_love

# 3이 출력되어야 합니다.
print(hot_cold(['냉정', '사랑', '사랑', '사랑', '열정', '사랑', '사랑']))
```

### 유전 암호


**운명을 찾아낸 둘이니까, DNA!**

DNA는 아데닌( A ), 구아닌( G ), 티민( T ), 사이토신( C )의 4가지 염기로 구성되어있습니다. 이 DNA는 mRNA에게 정보를 복사하여 주는데(=전사), 단순히 전달하는 것이 아닌, 어떠한 규칙에 의해 전달해줍니다. 이 규칙은 다음과 같습니다.

DNA의 염기&nbsp;&nbsp;&nbsp;&nbsp;	mRNA의 염기

---
A	&nbsp;&nbsp;&nbsp;&nbsp;U  
T	&nbsp;&nbsp;&nbsp;&nbsp;A  
G	&nbsp;&nbsp;&nbsp;&nbsp;C  
C	&nbsp;&nbsp;&nbsp;&nbsp;G  

서윤이는 어떠한 DNA의 염기 서열의 일부를 알고 있습니다. 이 염기 서열에서 mRNA로 얼마나 많은 아데닌(A)이 전사되는지 궁금합니다. 서윤을 도와 전사되어 생기는 아데닌의 개수를 구해봅시다.

 #string함수 #문자열 #반복문

**스스로 해봐요!**
- DNA의 염기 서열이 입력으로 주어질 때, 전사되는 mRNA의 아데닌(A)의 개수를 출력해주세요.

예시 입력

>TACAGAACAAAAATT

예시 출력

>3

예시 설명

주어진 DNA 염기 서열에서 전사되는 mRNA은 AUG UCU UGU UUU UAA 입니다. 여기서 아데닌의 개수는 3개이므로, 3을 출력합니다.

```
def count_adenine(dna_sequence):
    adenine_count = 0

    for base in dna_sequence:
        if base == 'T':
            adenine_count += 1

    return adenine_count

# 예시 입력
dna_sequence = "TACAGAACAAAAATT"

# 결과 출력
print(count_adenine(dna_sequence))
```
