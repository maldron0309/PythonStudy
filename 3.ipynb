### [실습5] 시퀀스의 인덱싱/슬라이싱


**시퀀스의 인덱싱과 슬라이싱**
- 인덱싱 : 시퀀스의 한 원소를 가져오는 것
- 슬라이싱 : 시퀀스의 일부분을 가져오는 것

>my_list = [1, 2, 3, 4, 5]  
print(my_list[2]) #3  
print(my_list[1:4]) #[2, 3, 4]  

아래 실습을 따라해보면서 시퀀스에서 인덱싱과 슬라이싱을 하는 방법을 익혀봅시다.

---

**이렇게 해봐요!**  
1. 문자열 “Impossible”이 담긴 변수 my_str을 만들어봅시다.

2. 리스트 ["Apple", "Banana", "Chamwae", "Durian"]이 차례대로 담긴 변수 my_list를 만들어봅시다.

3. 리스트 my_list에서 인덱싱을 이용해 변수 var1에 "Chamwae"를 넣어봅시다.

4. 문자열 my_str에서 슬라이싱을 이용해 변수 var2에 "possible"을 넣어봅시다.

**Tip!**
- 슬라이싱을 할 때 숫자를 적어주지 않을 수도 있어요! 다음 예제를 함께 살펴볼까요?

>my_list = [1, 2, 3, 4, 5, 6]  
print(my_list[3:]) # [4, 5, 6]  
print(my_list[:2]) # [1, 2]  
print(my_list[:]) # [1, 2, 3, 4, 5, 6]

- 첫번째 숫자를 생략하면 맨 앞에서, 두번째 숫자를 생략하면 맨 뒤까지 라는 의미를 가지고 있어요.

```
#문자열 “Impossible”이 담긴 변수 my_str을 만들어봅시다.
my_str = "Impossible"

#리스트 ["Apple", "Banana", "Chamwae", "Durian"]이 차례대로 담긴 변수 my_list를 만들어봅시다.
my_list = ["Apple", "Banana", "Chamwae", "Durian"]

#리스트 my_list에서 인덱싱을 이용해 변수 var1에 "Chamwae"를 넣어봅시다.
var1 = my_list[2]

#문자열 my_str에서 슬라이싱을 이용해 변수 var2에 "possible"을 넣어봅시다.
var2 = my_str[2:9]

print(var1)
print(var2)
```

### [실습6] 시퀀스의 길이/멤버 조사


**시퀀스의 길이와 멤버 조사하기**
- 길이 : len(시퀀스)
- 멤버 조사 : 원소 in 시퀀스 (존재하면 True, 존재하지 않으면 False)

>my_str = "abc"  
print(len(my_str)) #3 - my_str의 길이  
print('d' in my_str) #False - 존재하지 않음  

아래 실습을 따라해보면서 시퀀스의 길이와 멤버를 조사하는 방법을 익혀봅시다.

**이렇게 해봐요!**
1. 문자열 "Impossible"이 담긴 변수 my_str을 만들어봅시다.

2. 리스트 ["Apple", "Banana", "Chamwae", "Durian"]이 담긴 변수 my_list를 만들어봅시다.

1. len을 이용해서 변수 var1에 my_str의 길이를 넣어봅시다.

1. in을 이용해서 "Egg"가 my_list에 들어있는지 확인하고, 이 결과(True or False)를 변수 var2에 넣어봅시다.

```
#문자열 “Impossible”이 담긴 변수 my_str을 만들어봅시다.
my_str = "Impossible"

#리스트 ["Apple", "Banana", "Chamwae", "Durian"]이 차례대로 담긴 변수 my_list를 만들어봅시다.
my_list = ["Apple", "Banana", "Chamwae", "Durian"]

#len을 이용해서 변수 var1에 my_str의 길이를 넣어봅시다.
var1 = len(my_str)

#in을 이용해서 "Egg"가 my_list에 들어있는지 확인하고, 이 결과(True or False)를 변수 var2에 넣어봅시다.
var2 = "Egg" in my_list

print(var1)
print(var2)
```

### [실습7] 시퀀스의 연결/반복


**시퀀스의 연결과 반복**
- 연결 연산 : + - 두 시퀀스를 연결해줍니다(with 시퀀스)
- 반복 연산 : * - 한 시퀀스를 반복합니다(with 숫자)

>print( "안녕" + "하세요" ) #안녕하세요  
print( [1, 2] * 2 ) #[1, 2, 1, 2]

아래 실습을 따라해보면서 시퀀스를 연결하고 반복하는 방법을 익혀봅시다.

**이렇게 해봐요!**
1. 문자열 "Py"가 담긴 변수 str1과 문자열 "thon"이 담긴 변수 str2를 만들어봅시다.

1. 리스트 [3, 6, 9]가 담긴 변수 my_list를 만들어봅시다.

1. str1과 str2를 연결한 문자열을 변수 var1에 넣어봅시다

1. my_list를 3번 반복한 리스트를 변수 var2에 넣어봅시다.

```
# 문자열 "Py"가 담긴 변수 str1과 문자열 "thon"이 담긴 변수 str2를 만들어봅시다.
str1 = "Py"
str2 = "thon"

# 리스트 [3, 6, 9]가 담긴 변수 my_list를 만들어봅시다.
my_list = [3, 6, 9]

# str1과 str2를 연결한 문자열을 변수 var1에 넣어봅시다.
var1 = str1 + str2

# my_list를 3번 반복한 리스트를 변수 var2에 넣어봅시다.
var2 = my_list * 3

print(var1)
print(var2)

```

### [미션1] 무기 강화


**문자열로 무기를 강화해서 나만의 무기를 만들어봅시다!**

엘리스에서 최근 출시한 게임 금코토M에서는 다음과 같이 무기 강화를 진행합니다.

세 문자열 item1, item2, item3이 주어졌을 때, 이들을 이어붙인 문자열 + “무기” 형태의 무기 weapon이 완성됩니다.
예를 들어,

>item1 = "완전 좋고"  
item2 = "빛나며"  
item3 = "손에 착착 감기는"  

의 값을 대입하여,

>완전 좋고 빛나며 손에 착착 감기는 무기  

의 결과를 출력할 수 있습니다.

변수와 시퀀스의 연결 연산을 이용해서 여러분만의 무기를 만들어봅시다!

---
**이렇게 해봐요!**
1. 변수 item1, item2, item3을 선언하고, 문자열을 자유롭게 넣어봅시다.

2. item1, item2, item3과 “무기” 문자열을 차례대로 이어 붙여서 여러분의 무기 weapon을 만들어봅시다.

3. 만든 무기 weapon을 출력해봅시다.

weapon은 아래와 같은 형식이어야 합니다.

>item1 item2 item3 무기  
(예 : 완전 좋고 빛나며 손에 착착 감기는 무기)


**주의사항**  
weapon에서 각 item 사이에는 공백이 들어가있으므로,

> itemA + " " + itemB   

와 같은 형태로 공백 문자열을 이용해서 공백을 삽입해 주어야 합니다.

```
# 변수 item1, item2, item3을 선언하고 문자열을 자유롭게 넣어봅시다.
item1 = "완전 좋고"
item2 = "빛나며"
item3 = "손에 착착 감기는"

# 무기 weapon을 만들어봅시다.
weapon = item1 + " " + item2 + " " + item3 + " 무기"

# 만든 무기 weapon을 출력해봅시다.
print(weapon)
```

### [미션2] "똑똑한" .remove() 사용법


**remove를 잘 쓰려면 우선 있는지부터 확인!**  
리스트에서 .remove()는 원소를 삭제하는 데에 쓰이지만, 만약 그 원소가 리스트 내부에 없을 경우, 오류가 발생합니다.  

따라서 우리는 지우려는 원소가 리스트에 있는지 확인한 다음, 원소를 지우거나 다른 조치를 취해야 합니다.  

아래의 절차를 따라서 리스트 안의 원소의 값을 오류 없이 제거해봅시다.  

---
**이렇게 해봐요!**  
다음 과정을 통해 "Durian"과 "Pineapple"을 지워봅시다.
*   원소가 있다면 : 해당 원소를 리스트에서 지웁니다.  
*   원소가 없다면 : 해당 원소가 리스트 안에 없다는 문장을 출력합니다. (“원소은(는) 리스트 안에 없습니다!”)

```
# 과일들이 담긴 리스트 fruits입니다.
fruits = ['Apple', 'Banana', 'Chamwae', 'Durian']

# "Durian"을 fruits에서 지워봅시다.
fruits.remove('Durian')

# "Pineapple"을 fruits에서 지워봅시다. (만약 "Pineapple"이 fruits에 없으면 에러 발생)
fruits.remove('Pineapple')
```
