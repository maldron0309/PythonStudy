### 2. 우뚝 솟은 빌딩! 함께 만들어볼까요?  
입력으로 숫자 N이 주어지면, 다음과 같은 규칙으로 빌딩을 출력하는 코드를 작성해봅시다.

nn이 1~5인 경우, nn층 만큼의 계단 별자리를 출력합니다.  
nn이 5 이상이면, 1과 더불어 n ≥ 5n≥5인 구간에서는 별을 항상 5개 출력합니다.  
예시 입력 1  
3  

예시 출력 1  
@  
@@  
@@@


예시 입력 2  
7  

예시 출력 2  
@   
@@  
@@@  
@@@@  
@@@@@  
@@@@@  
@@@@@  

**풀어보세요!**
숫자 입력이 주어질 때, 주어진 규칙에 맞는 빌딩을 출력해봅시다.

```
# 숫자 입력이 주어질 때, 주어진 규칙에 맞는 빌딩을 출력해봅시다.
N = int(input("숫자를 입력하세요: "))

for i in range(1, N + 1):
    if i >= 5:
        print("@" * 5)
    else:
        print("@" * i)
```

### 3. 369 369! 369 369!**
승희와 준배, 그리고 호준은 369 게임을 하려고 합니다. 코더랜드에서는 3, 6, 9 게임의 룰이 다음과 같다고 합니다.

1. 한사람씩 순서대로 1부터 숫자를 하나씩 이야기합니다. (1, 2, 3, …)  
2. 만약 자기가 이야기해야할 수가 3의 배수라면 (3, 6, 9, 12, …) 박수를 칩니다.   

호준이는 수학을 잘 못하기 때문에 이 룰을 이해하지 못하고 있습니다. 호준이가 3, 6, 9 게임에 걸리지 않게 여러분들이 이 프로그램을 작성해서 도와줍시다.   

예시 입력 1  
7  

예시 출력 1  
8  

예시 입력 2  
11  
  
예시 출력 2  
짝!  

**풀어보세요!**
수를 하나 입력받아서, 그 다음 수의 사람이 어떤 동작을 해야할지 출력하는 프로그램을 작성해봅시다.

* 숫자를 이야기하는 경우 : 그 숫자를 출력합니다.
* 박수를 쳐야하는 경우 : 짝!을 출력합니다.

```
number = int(input("숫자를 입력하세요: "))

for i in range(1, number + 1):
    print(i)

    if i % 3 == 0:
        print("짝!")
```

### 4. 어서오세요! 커피전문점 수타박수입니다!
세계에서 가장 맛있는 커피를 파는 수타박수에서는 매일 많은 사람들이 커피를 마시기 위해 방문하는 장소입니다. 수타박수에서 일하는 직원 엘리스는 너무 많은 손님이 와서 골머리를 앓고 있습니다.

특히, 엘리스는 많은 손님들이 오면 더욱 어려움을 느낍니다. 주문이 한번에 5개, 10개 등이 들어왔을 때 각 주문을 외우는 것이 어렵기 때문입니다. 이런 어려움을 겪는 엘리스를 돕기 위해 주문을 입력하면 총 금액을 출력하는 프로그램을 작성해봅시다.

-수타박수의 메뉴표-  

Coffee	Price  
아메리카노	4100  
카페라떼	4600  
카라멜마끼아또	5100  

예시 입력 1  
3  
아메리카노  
카페라떼  
카라멜마끼아또  
  
예시 출력 1    
13800    
  
**풀어보세요!**
입력  
첫번째 줄에 숫자 하나(n)를 입력받습니다.  
두번째 줄부터 n+1번째 줄까지 메뉴(아메리카노, 카페라떼, 카라멜마끼아또)중 하나를 입력받습니다.  

출력  
주문한 메뉴들의 총 가격을 출력합니다.  

```
menu_prices = {
    '아메리카노': 4100,
    '카페라떼': 4600,
    '카라멜마끼아또': 5100
}

n = int(input())

total_price = 0
for _ in range(n):
    menu = input()
    total_price += menu_prices.get(menu, 0)

print(total_price)

```

### 5. 이번 교수님은 만만치않다…!
코더랜드에 있는 F대학의 K교수님은 학점을 특이하게 주시는 것으로 유명합니다.

이 교수님의 시험은 3가지 과목 - 언어이해, 추리논증, 논술로 구성됩니다. 세 과목의 시험 점수 중 가장 낮은 점수에 대해 50점 이상이면 학생에게 A를 주시고, 75점 이상이면 A+를 부여합니다. (각 시험은 100점 만점 시험입니다.)

단, 이 교수님은 추리논증부분에서 10년 넘게 공부하셨기 때문에, 이 과목에 성의가 없는 학생을 싫어하셔서
추리논증 점수가 50점이 넘지 않으면 F를 부여합니다.

위 경우에 모두 해당하지 않는 학생들에게는 전부 B+를 부여합니다.

어떤 학생이 받을 학점을 출력해주는 프로그램을 작성해봅시다.

입력 예시 1  
70 75 80   
'# 이때 70은 언어이해 점수  
'# 75는 추리논증 점수  
'# 80은 논술 점수입니다.  

출력 예시 1  
A  

입력 예시 2  
95 45 90  

출력 예시 2  
F  

**풀어보세요!**  
* 어떤 학생의 세 과목 점수가 입력 예시의 형식대로   - 언어이해 추리논증 논술 순으로 -   
띄어쓰기로 구분해서 주어질 때 이 학생이 받을 학점을 출력해봅시다.

* 변수 scores에는 입력하는 순서대로 [언어이해, 추리논증, 논술] 과 같이 리스트의 형태로 점수가 담겨있으니 이를 활용해서 문제를 풀어봅시다.

```

scores = list(map(int, input().split()))

min_score = min(scores)

if min_score >= 75:
    grade = 'A+'
elif min_score >= 50:
    grade = 'A'
elif scores[1] < 50:
    grade = 'F'
else:
    grade = 'B+'

print(grade)
```
