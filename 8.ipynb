### [실습3] OOP의 특징 : 상속


**상속하기**

객체지향 프로그래밍에서는 여러 특징이 존재하는데, 이번 Vol.2에서는 2가지를 먼저 다루어보고자 합니다.

- 상속 : 한 클래스의 내용을 다른 클래스가 이어받는 것

>class Car:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    type = "자동차"  
>
>class Hyundai(Car):  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    pass
>
>sonata = Hyundai()  
print(sonata.type) #자동차  

상속을 통해 한 클래스에서 다른 클래스에게 핵심 정보를 전달해줄 수 있습니다.

**이렇게 해봐요!**
1. Pet 클래스를 만들어보고, 다음 변수와 메서드를 추가해봅시다.
- 변수 attr = “cute”
- 메서드 cry = “멍멍!”을 반환
2. Dog 클래스를 만들어보고, Pet을 상속받아봅시다.

3. Dog 클래스로 인스턴스 maltese를 만들어봅시다.

**Tips**  
- pass는 아무런 동작도 하지 않는 코드입니다.

- 함수를 만들고 아무런 코드도 없다면 오류를 일으키기 때문에 무의미한 pass를 넣습니다.

```
class Pet:
    attr = "cute"
    def cry(self):
        return "멍멍!"
class Dog(Pet):
    pass

maltese = Dog()
```

### [실습4] OOP의 특징 : 다형성


**다형성**
- 다형성 : 같은 모양의 코드가 다른 역할을 하는 것  
>class Car:
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    type = "자동차"  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    horsePower = 100  
>
>class Lamborghini(Car):  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    horsePower = 700
>
aventador = Lamborghini()  
print(aventador.type) #자동차  
print(aventador.horsePower) #700  
  
다형성을 통해 유사한 역할을 하는 변수와 메서드를 같은 이름으로 관리할 수 있습니다.

**이렇게 해봐요!**
1. Animal 클래스를 만들어보고, 다음 변수와 메서드를 추가해봅시다.
- 변수 legs = 0
- 메서드 walk : 빈 문자열(“”)을 반환
2. Dog 클래스를 만들어봅시다. Animal 클래스를 상속받고, 변수 legs를 4로, 메서드 walk는 “살랑살랑”을 반환하도록코드를 작성해봅시다.

3. Human 클래스를 만들어봅시다. Animal 클래스를 상속받고, 변수 legs를 2로, 메서드 walk는 “뚜벅뚜벅”을 반환하도록코드를 작성해봅시다.

4. Dog 클래스의 인스턴스인 maltese, Human 클래스의 인스턴스인 gildong을 만들어봅시다.

```
class Animal:
    legs = 0
    def walk(self):
        return ""
class Dog(Animal):
    legs = 4
    def walk(self):
        return "살랑살랑"
class Human(Animal):
    legs = 2
    def walk(self):
        return "뚜벅뚜벅"
maltese = Dog()
gildong = Human()
```

겨울만 되면 어김없이 주황색 천막에 가려진 소소한 즐거움이 있습니다. 바로 붕어빵 가게죠!

붕어빵을 만드는 과정은 붕어빵틀에 붕어빵의 재료를 넣고, 앞뒤로 잘 구워주면 맛있는 붕어빵이 나옵니다.

클래스와 인스턴스의 관계도 이와 비슷한데요, 붕어빵을 만드는 틀(클래스)가 있다면, 이로부터 붕어빵(인스턴스)를 만들 수 있습니다!

오른쪽 미션을 통해 붕어빵 클래스를 생성하고, 이로부터 팥 붕어빵, 슈크림 붕어빵을 만들어봅시다!

---
**미션! 해결해봅시다.**
1~3, 9, 13, 16, 19, 22번째 줄의 미션을 확인하고, 이를 수행해보세요.

**Tip!**

- 한 객체의 메서드에서 그 객체의 필드를 참조할 땐 self를 이용해서 참조해야 합니다.
- 클래스를 만드는방법과 인스턴스를 만드는 방법은 다르죠! 이를 구분하면서 작성해봅시다.

예시)

>class Human: # Human 클래스 생성!  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    age = 10  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    def talk(self):

>
>gildong = Human() # Human 클래스의 인스턴스 gildong 생성!  
gildong.talk() #10 살입니다!

```
# 클래스 Bbread(붕어빵)을 만들어봅시다.
class Bbread:
    # Bbread의 필드 taste를 생성하고, "밀가루"를 대입해봅시다.
    taste = "밀가루"

    # Bbread의 메서드 eat()를 생성해봅시다. 이 메서드는 "taste 맛이 나요!"를 출력합니다.
    def eat(self):
        print(f"{self.taste} 맛이 나요!")

# 클래스 Bbread의 인스턴스 redBean과 choux를 만들어봅시다.
redBean = Bbread()
choux = Bbread()

# 객체 redBean의 필드 taste를 "팥"으로 바꿔줍시다
redBean.taste = "팥"

# 객체 choux의 필드 taste를 "슈크림"으로 바꿔줍시다
choux.taste = "슈크림"

# 객체 redBean의 메서드 eat()를 실행해봅시다
redBean.eat()

# 객체 choux의 메서드 eat()를 실행해봅시다
choux.eat()
```
