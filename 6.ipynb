###[실습1] 내장함수 사용 : 최댓값/최솟값

**최댓값/최솟값 쉽게 구하기**  
시퀀스 안의 원소 중, 최댓값과 최솟값을 구하는 함수는 다음과 같습니다.  

- 최댓값 : max(시퀀스)
- 최솟값 : min(시퀀스)  

단, 시퀀스 안의 원소가 모두 숫자형 자료여야 합니다!

>my_list = [10, 20, 30, 40, 50]
>
>print(max(my_list)) #50
print(min(my_list)) #10

아래 실습을 따라해보면서 최댓값과 최솟값을 쉽게 구하는 방법을 익혀봅시다.

**이렇게 해봐요!**  
1. 원하는 숫자 5개가 담긴 튜플 my_tuple을 선언해봅시다.

1. 변수 var1를 선언하고, my_tuple의 최댓값을 넣어봅시다.

1. 변수 var2를 선언하고, my_tuple의 최솟값을 넣어봅시다.

**Tip!**  
시퀀스 중 문자열은 원소가 모두 문자열이므로 max(), min()을 사용할 수 없겠죠?

```
#원하는 숫자 5개가 담긴 튜플 my_tuple을 선언해봅시다.
my_tuple = (25, 13, 45, 7, 32)

#변수 var1를 선언하고, my_tuple의 최댓값을 넣어봅시다.
var1 = max(my_tuple)

#변수 var2를 선언하고, my_tuple의 최솟값을 넣어봅시다.
var2 = min(my_tuple)

#결과 출력
print("최댓값:", var1)
print("최솟값:", var2)
```


###[실습7] 최댓값 직접 구하기

**내장 함수를 사용할 수 없다면? 직접 함수를 만들어봐요!**  
이전 실습에서는 max()를 이용해서 최댓값을 구했었는데요, 이번 실습에서는 직접 최댓값을 구하는 함수 our_max() 함수를 구현해봅시다!  

**이렇게 해봐요!**
max()를 사용하지 않고, 어떤 리스트를 인자로 넘겨주면 그 리스트의 최댓값을 반환(return) 하는 함수 our_max()를 만들어봅시다.

```
# 리스트 nums를 넣었을 때, 최댓값을 반환(return)하는 함수 our_max를 작성해봅시다.
def our_max(nums):
    # 리스트가 비어있으면 None을 반환
    if not nums:
        return None

    # 최댓값 초기화
    maximum = nums[0]

    # 리스트의 모든 요소와 최댓값을 비교하며 갱신
    for num in nums:
        if num > maximum:
            maximum = num

    return maximum

# 결과 출력
print(our_max([1, 2, 10, 9, 3, 7, 0, 99, 27, 85]))
```

###[미션 1] 네버랜드

**환상의 나라로!**
유진이는 국내 최고의 도시 용인에 있는 네버랜드에 도착했습니다. 네버랜드에는 정말 많은 놀이기구가 있어서 유진이는 특별한 순서로 놀이기구를 타기로 결정했습니다.

각 놀이기구의 대기시간이 담긴 리스트가 인자로 주어졌을 때, 유진이가 놀이기구를 타는 순서로 대기시간을 정렬한 함수 neverland()를 만들어봅시다.

**이렇게 해봐요!**  
대기시간을 다음의 조건에 맞게 정렬하는 함수 neverland()를 작성해봅시다.

- 대기시간이 가장 짧은 놀이기구부터 오름차순으로 놀이기구를 탑니다.

- 단, 인덱스 2에 해당하는 놀이기구는 유진이가 꼭 타고 싶어하는 놀이기구이기에 대기시간에 상관없이 가장 먼저 탑니다.

- 인자 : 리스트 queue

- 반환값 : 주어진 조건에 맞게 정렬된 리스트

**예시 1**

>queue = [0,10,100]  
print(neverland(queue))# [100,0,10]
  
**예시 2**

>queue = [10,20,30,40,50]  
print(neverland(queue)) # [30,10,20,40,50]

**Hint!**  
리스트의 메서드 중에는 .sort()가 있습니다. 이는 숫자들을 오름차순으로 정렬해줍니다.

>nums = [20, 10, 30, 50, 40]  
nums.sort() #반환값이 없음!  
print(nums) # [10, 20, 30, 40, 50]  

**유의사항**

- 대기시간이 같은 놀이기구는 없음

```
# 대기시간이 담긴 리스트가 인자로 주어지면 조건을 만족하도록
# 타야하는 대기시간의 순서가 담긴 리스트를 반환하는 함수 neverland()를 작성해봅시다.
def neverland(queue):
    # 인덱스 2에 해당하는 놀이기구는 가장 먼저 타야함
    first_ride = queue[2]

    # 대기시간이 가장 짧은 놀이기구부터 오름차순으로 정렬
    queue.sort()

    # 정렬된 리스트에서 인덱스 2에 해당하는 놀이기구를 가장 앞으로 이동
    queue.remove(first_ride)
    queue.insert(0, first_ride)

    return queue

# 확인을 위한 코드입니다.
# 대기시간이 담긴 리스트 queue를 자유롭게 수정해보세요!
queue = [30, 10, 20, 50, 40, 60]
print(neverland(queue))
```

**하고싶은 대로만 한다면 다 된다고 믿어!**

엘리스 홈페이지의 비밀번호를 어떻게 설정할지 고민이 많던 건웅은 듣고 있던 노래 윤하 - 비밀번호 486에서 영감을 받아 다음과 같이 비밀번호를 만들기로 했습니다.

4, 6, 8로 이루어진 숫자

숫자	문자열

---
4	  love  
8	  smile  
6	  kiss  

예를 들어 다음과 같은 숫자가 입력으로 주어지면

>48686
  
다음과 같은 문자열 비밀번호를 반환합니다.

>lovesmilekisssmilekiss


**이렇게 해봐요!**  
주어진 조건에 맞는 비밀번호를 만들어주는 함수 yoonHa()를 만들어봅시다.

- 인자 : 숫자 nums
- 반환값 : 변환한 비밀번호 문자열

예시 입력
>486

예시 출력
>lovesmilekiss
  
**Hints!**  
1. 문자열의 각 글자를 반복 인자로 사용할 수 있어요.

>for i in "486":  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;print(i)

2. 문자열에 연산자를 사용해 간단하게 문자열을 붙힐 수 있어요.

>str1 = ""  
str2 = "love"  
str1 = str2 + str2 + str2  
print(str1)  

3. 사전형 자료형을 활용하면 더욱 쉽게 문제를 해결할 수 있어요.

4. 입력되는 수는 4, 8, 6 만 있어요.

5. 예시로 주어진 486은 예시예요. 모든 4, 8, 6의 조합에 대해 만족하도록 코드를 작성해주세요.

```
# 비밀번호를 만들어주는 함수 yoonHa()를 만들어봅시다.
def yoonHa(nums):
    # 숫자에 대응하는 문자열 사전
    mapping = {'4': 'love', '8': 'smile', '6': 'kiss'}

    # 비밀번호 문자열 초기화
    password = ""

    # 입력된 숫자의 각 자리에 대응하는 문자열을 비밀번호에 추가
    for digit in str(nums):
        password += mapping[digit]

    return password

# 채점을 위한 코드입니다. 이를 수정하지 마세요!
nums = int(input())
print(yoonHa(nums))
```
