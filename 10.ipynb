### 단순 캘린더


**오늘 일정 있는 사람, 나야나!**

달력을 활용하면 더욱 효율적으로 일정 관리를 할 수 있습니다. 이런 달력의 특징들을 담은 클래스를 만들어둔다면, 달력 인스턴스를 여러 개 생성해서 편하게 달력을 활용할 수 있겠죠?

우선 간단하게 아래 기능이 담긴 클래스 Calendar()을 만들어봅시다.

 #클래스

---
**스스로 해봐요!**

다음 기능을 가진 캘린더 클래스 Calendar()를 만들어봅시다.

- 필드 year : “년”을 나타내는 변수

- 필드 month : “월”을 나타내는 변수

- 필드 day : “일”을 나타내는 변수

- 메서드 today() : 오늘 날짜를 “0000년 0월 0일” 꼴의 문자열로 반환합니다.

- 메서드 set_day(a, b, c) : 달력의 날짜를 a년 b월 c일로 설정합니다. - 이때 a, b, c는 숫자

```
class Calendar():
    year = 0
    month = 0
    day = 0

    # today 메서드를 구현합니다.
    def today(self):
        return f"{self.year}년 {self.month}월 {self.day}일"

    # set_day 메서드를 구현합니다.
    def set_day(self, y, m, d):
        self.year = y
        self.month = m
        self.day = d

# Calendar 클래스의 인스턴스 생성
siri = Calendar()

# set_day 메서드를 사용하여 날짜 설정
siri.set_day(2019, 7, 18)

# today 메서드를 사용하여 오늘 날짜 출력
print(siri.today())
```

### 뒤죽박죽 덧셈


**숫자로 연산자를 더해봅시다.**
코더랜드에는 숫자로 연산자를 더하는 뒤죽박죽 덧셈이 있습니다. 아래 규칙에 따라 이를 진행하는 코드를 작성해봅시다.

예시 입력 1

>+4-

>++++----

>*10/

 #시퀸스의_연결 #조건문

---
**스스로 해봐요!**  
1. 다음과 같이 입력이 주어질 때,  
(연산자1)N(연산자2) - 이때 N은 숫자

2. 연산자1과 연산자2를 뒤죽박죽 덧셈 해봅시다.

3. 이때, 뒤죽박죽 덧셈은 다음과 같습니다 :
- N이 5의 배수이면 연산자1만 N번만큼 출력합니다.
- N이 3의 배수이면 연산자2만 N번만큼 출력합니다.
- 만약 N이 3의 배수이면서 5의 배수이면 아무것도 출력하지 않습니다.
- 이외의 경우에는 두 연산자를 각각 N회만큼 출력해줍니다.

```
s = input()

o1 = s[0]
o2 = s[-1]

num = int(s[1:-1])  # 연산자가 아닌 숫자 부분

result = ""

# N이 5의 배수이면 연산자1만 N번만큼 출력
if num % 5 == 0:
    result = o1 * num
# N이 3의 배수이면 연산자2만 N번만큼 출력
elif num % 3 == 0:
    result = o2 * num
# N이 3의 배수이면서 5의 배수이면 아무것도 출력하지 않음
elif num % 3 == 0 and num % 5 == 0:
    pass
# 이외의 경우에는 두 연산자를 각각 N회만큼 출력
else:
    result = o1 * num + o2 * num

print(result)
```

### 끼리끼리


**끼리끼리**

지금까지 우리는 sort() 함수를 이용해서 정렬을 진행하였습니다. 그렇다면 다른 방법으로 정렬하는 방법은 없을까요? 다음의 정렬을 직접 한번 구현해봅시다.

형운이는 숫자를 정렬하고자 합니다. 정수(양수 혹은 음수)를 공백( )을 기준으로 입력받았을 때 음수는 앞쪽에 분류를, 양수는 뒷쪽에 분류를 하려고 합니다. 형운이를 도와 문제를 해결해봅시다.

예시 입력
>-1 2 1 3 -2

예시 출력
>-1 -2 2 1 3

 #리스트 #삽입

---
**스스로 해봐요!**.
- 여러 정수를 공백( )을 기준으로 입력을 받습니다.
- 이 입력은 주어진 코드에 의해 리스트 자료로 변환되어 변수 my_list에 담깁니다.
- 음수를 양수보다 앞에, 양수는 음수보다 뒤에 출력을 해봅시다. (0은 양수로 취급합니다.)
- 단, 같은 부호의 숫자끼리의 순서에는 변함이 없어야 합니다.

**Hint!**  
append 함수는 리스트의 앞부분부터 삽입을 하는 함수입니다.

```
def my_sort(lis):
    # 음수와 양수를 분리
    negatives = [x for x in lis if x < 0]
    positives = [x for x in lis if x >= 0]

    # 음수 리스트를 뒤집어서 양수 리스트 뒤에 추가
    negatives.reverse()
    result = negatives + positives

    return result

# 공백을 기준으로 입력받고 함수를 수행하는 코드입니다. 수정하지 마세요!
my_list = [int(v) for v in input().split()]
print(my_sort(my_list))
```

### 이상한 나라의 언어

**이상한 나라의 언어**

엘리스가 떨어진 이상한 나라에서는 특이한 언어 방식을 채택합니다. 한 문장 안에 단어들을 거꾸로 말하는 방식입니다.

예를 들어 “안녕 엘리스”를 이들은 “엘리스 안녕” 이라고 말을 합니다.

엘리스는 이제 막 떨어져서 그들의 말을 알아듣기 힘들어합니다. 엘리스를 도와 단어를 바꿔주는 통역 프로그램을 만들어주세요!

#리스트 #문자열

---
**스스로 해봐요!**
- 첫째 줄에 문장을 입력합니다. 단어와 단어 사이에는 공백 하나만 있습니다.

- 입력받은 문장을 뒤집어서 출력합니다.

입력 예시

>엘리스 안녕!

출력 예시

>안녕! 엘리스

**Hint!**  
리스트 자료형에는 .reverse() 메서드가 있습니다. 이는 리스트를 뒤집어줍니다.

>a = [1, 2, 3]  
a.reverse()  
print(a)  # [3, 2, 1]

```
def reverseWords(s):
    words = s.split()
    words.reverse()
    result = ' '.join(words)
    return result

line = input()
print(reverseWords(line))
```
